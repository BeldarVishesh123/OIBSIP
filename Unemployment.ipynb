```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df1 = pd.read_csv("Unemployment_Rate_upto_11_2020.csv")
df2 = pd.read_csv("Unemployment in India.csv")
df1.columns = df1.columns.str.strip()
df2.columns = df2.columns.str.strip()
df1["Date"] = pd.to_datetime(df1["Date"], errors="coerce", dayfirst=True)
df2["Date"] = pd.to_datetime(df2["Date"], errors="coerce", dayfirst=True)
print("Dataset 1 Shape:", df1.shape)
print("Dataset 2 Shape:", df2.shape)
print("\nDataset 1 Columns:", df1.columns)
print("Dataset 2 Columns:", df2.columns)
print("\nFirst rows of Dataset 1:\n", df1.head())
print("\nFirst rows of Dataset 2:\n", df2.head())
print("\nMissing values in Dataset 1:\n", df1.isnull().sum())
print("\nMissing values in Dataset 2:\n", df2.isnull().sum())
df1.dropna(subset=["Date", "Region"], inplace=True)
df2.dropna(subset=["Date", "Region"], inplace=True)
print("\nSummary of Dataset 1:\n", df1.describe(include="all"))
print("\nSummary of Dataset 2:\n", df2.describe(include="all"))
plt.figure(figsize=(12,6))
sns.lineplot(data=df1, x="Date", y="Estimated Unemployment Rate (%)", label="Dataset 1")
sns.lineplot(data=df2, x="Date", y="Estimated Unemployment Rate (%)", label="Dataset 2")
plt.title("Unemployment Rate Over Time")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.legend()
plt.show()
plt.figure(figsize=(12,6))
sns.boxplot(data=df1, x="Region.1", y="Estimated Unemployment Rate (%)")
plt.xticks(rotation=90)
plt.title("Unemployment Rate Distribution by Region (Dataset 1)")
plt.show()
plt.figure(figsize=(12,6))
sns.boxplot(data=df2, x="Area", y="Estimated Unemployment Rate (%)")
plt.title("Unemployment Rate Distribution by Area (Dataset 2)")
plt.show()
plt.figure(figsize=(8,6))
sns.heatmap(df1.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap (Dataset 1)")
plt.show()
plt.figure(figsize=(8,6))
sns.heatmap(df2.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap (Dataset 2)")
plt.show()
```
